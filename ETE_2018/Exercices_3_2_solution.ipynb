## a)
créez un vecteur `x` dans lequel on trouve 300 v.a $x_i$ tirées d'une loi uniforme $x_i \sim U\left[-2, 2\right]$


```r
x <- runif(n = 300, min = -2, max = 2)
```

## b)
créez un vecteur `beta` dans lequel on trouve 300 v.a $\beta_i$ tirées d'une loi normale $\beta_i \sim N(0,1)$


```r
beta <- rnorm(n = 300)
```

## c)
créez un vecteur `y` ayant la forme suivante:
    $$y_i=x_i^3+\beta_i$$
    

```r
y <- x**3+beta
```

## d)
faites un graphique de $y$ par rapport $x$

```r
plot(x, y)
```

## e)
Créez une fonction `graph_1` qui un graphique $y$ en fonction de $x$ de la question précédente. Toutefois, ajouter l'argument `nb=...`. Cet argument permet de fixer le nombre de variables aléatoires tirées.

Par exemple, lorsque j'exécute la fonction `graph_1(nb=1000)` j'obtiens le graphique suivant:

```r
graph_1 <- function(nb=300){
    x <- runif(n = nb, min = -2, max = 2)
    beta <- rnorm(n = nb)
    y <- x**3+beta
    return(plot(x, y))
}
```

```r
graph_1(nb=1000)
```

## f)
Ajouter à cette fonction l'argument `limMin=...` qui permet de fixer la limite inférieure, et `limMax=...` qui permet de fixer la limite supérieure de notre loi Uniforme d'où les $x_i$ ont été tirées.

Mais **attention!**, ajoutez un message d'erreur si la limite `limMin` n'est pas inférieure à la `limMax`. 

```r
graph_1 <- function(nb=300, limMin=-2, limMax=2){
    if(limMin<limMax){
        x <- runif(n = nb, min = limMin, max = limMax)
        beta <- rnorm(n = nb)
        y <- x**3+beta
        return(plot(x, y))
    }else{
        print("La limite limMin ne peut être supérieure ou égale à la limMax")
    }
}
```

```r
graph_1(nb=1000,limMin=-4, limMax=10)
```

```r
graph_1(nb=1000,limMin=4, limMax=2)
```

# Question 2
Supposons que vous voulez aider votre ancien professeur dans sa tâche d’assigner les résultats aux étudiants. Ainsi, le professeur n’a qu’à saisir un vecteur contenant les notes dans l’argument d’une fonction que vous lui concevez. Votre fonction lui retourne un _data frame_ avec la **noteObtenue** et le **résultat** de chaque étudiant.

En suivant l’intervalle des notes dans l’image ci-dessous, créer un script `R` qui vous permet d’obtenir le résultat (en lettre) que chaque étudiant reçoit dans son bulletin.

![test](https://raw.githubusercontent.com/nmeraihi/data/master/images/intervallesDesNotes.png)

```r
resultat <- function(notes=a){
    library(dplyr)
    resultat=case_when(
        notes>=90 & notes<=100 ~"A+",
        notes>=85 & notes<90 ~"A",
        notes>=80 & notes<85 ~"A-",
        notes>=76 & notes<80 ~"B+",
        notes>=73 & notes<76 ~"B",
        notes>=70 & notes<73 ~"B-",
        notes>=66 & notes<70 ~"C+",
        notes>=63 & notes<66 ~"C",
        notes>=60 & notes<63 ~"C-",
        notes>=55 & notes<60 ~"D+",
        notes>=50 & notes<55 ~"D",
        notes>=0 & notes<50 ~"E")
    df <- data.frame( notes, resultat)
    colnames(df) <- c("noteObtenue", "résultat")
    return(df)
}
```

Vous pouvez test votre script avec le vecteur `a` suivant 

```r
set.seed(123)
a <- round(runif(10,min = 60, max = 99),0)
```

```r
resultat(notes=a)
```

```r
set.seed(123)
b <- round(runif(100,min = 60, max = 99),0)
```

```r
resultat(notes=b)
```
