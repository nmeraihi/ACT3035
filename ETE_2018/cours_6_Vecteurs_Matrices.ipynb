## Création des vecteurs et matrices

```sas
proc iml;

title ;
x={1 2 3}; /* Vecteur (ligne) */
y={1 2 3, 4 5 6}; /* Une matrice numérique 2x3 */
z={"Homme", "Femme"}; /* Une autre matrice de type charactère 2x1 */
print x y z;

n_z=nrow(z); /* Le nombre de ligne du vecteur x */
n_y=nrow(y); /* Le nombre de la matrice y */
print n_z n_y ;


yy=y+y; /* La somme de la matrice y avec elle même */
print yy;

type_y=type(y); /* Le type de la matrice y */
type_z=type(z); /* Le type de la matrice z */
print type_y type_z;

quit;
```

## Créer des matrices constantes

```sas
proc iml;
c = j(10, 1, 3.14); /* Un vecteur colonne 10x1 */
r = j( 1, 5); /* Un vecteur ligne 1x5 de valeur 1 */
m = j(10, 5, 0); /* Une matrice 10 x 5 avec des 0*/
miss = j(3, 2, .); /* Une matrice 3 x 2 avec des valeurs manquantes*/

print c r m miss;
quit;
```

## Combiner des vecteurs et des matrices

```sas
/* concatenation horizontal*/
proc iml;
a = {1,2,3,4,5}; /* Un vecteur colonne 5 x 1 */
b = {3,5,4,1,3};
c = {0,1,0,0,1};
x = a || b || c; /* Ce qui donne une matrice 5x3 */
print a b c;

print x;


A = {1 2 3};
B = {4 5 6};
AHB = a || b;
AVB = a // b;

print A B;

print AHB AVB;

quit;
```

## Multiplication de matrices

```sas
proc iml;
A = {1 2, 3 4};
Asquared = A * A;
print A, A, Asquared;
quit;
```

## Calcul du déterminant

```sas
proc iml;
A = {1 2, 3 4};
detval = det(A);
print detval;
quit;
```

## Créer un ensemble de données à partir d'un ensemble de vecteurs et de matrices

```sas
proc iml;
y = {1,0,3,2,0,3}; /** vecteur 6 x 1 **/
z = {8,7,6,5,6};   /** vecteur 5 x 1 **/
c = {A A A B B B}; /** vecteur 1 x 6 de type caractère**/
 
create MyData var {y z c}; /** créer le data set **/
append;       /** Insérer les vecteurs dans le data set **/
close MyData; /** Fermer le data set **/
quit;

proc print data=MyData noobs;
quit;
```

un autre exemple avec une matrice 4x3

```sas

proc iml;
x = {1 2 3,
     4 5 6,
     7 8 9,
     3 2 1 }; 
create MyData2 from x[colname={"q" "r" "s"}];
append from x;
close MyData2;
quit;
proc print data=MyData2 noobs;
quit;
```

## Créer un vecteur avec des valeurs séquentielles

```sas
proc iml;
i=1:5; /* incrément de 1 */
j=5:1; /* incrément de -1 */
k=do(1, 10, 2); /* des chiffres inpairs 1, 3, ..., 9 */
i1=10;
i2=15;
inc=.75;
ex=do(i1, i2, inc);
print i j k ex;
quit;
```

## Transposer des matrices

```sas
proc iml;
A={1 2, 3 4};
trans_A=t(A); /* transposer avec la fonction t() */
trans_A2=A`; /* transposer avec l'accent grave' */
print A trans_A trans_A2;
quit;
```

## Changer le format de matrice

```sas
proc iml;
x=1:12;/* 1x12 */
s=shape(x, 4, 3);/* 4x3 */
print x s;
quit;
```

## Extraire une partie de la matrice

```sas
proc iml;
y={1 2 3, 4 5 6};
x=y[2,3];
y[2,3]=7;
print x y;
quit;
```

```sas
proc iml;
y={1 2 3 14, 4 5 6 16, 7 8 9 19};
rows={1 3};
cols={2 4};
w=y[rows, cols];
oddRows=y[rows, ]; /* Extraire toutes les lignes impaires */
evenCols=y[, cols]; /* Extraire toutes les lignes paires */

print y w oddRows evenCols;
quit;
```

## Extraire la diagonale d'une matrice

```sas
proc iml;
x={	1 2 3,
	4 5 6,
	3 2 1};
d=vecdiag(x);
print x;
print d;
quit;
```

## Modifier la diagonale d'une matrice

```sas
proc iml;
x={	1 2 3,
	4 5 6,
	3 2 1};
vals={5, 2, -1};
s=diag(vals);
print x;
print s;
quit;
```

## Créer une matrice identité

## Afficher une partie d'une matrice avec noms de colonnes et étiquette

```sas
proc iml;
x=shape(1:1000, 0, 4);
varNames="Col1":"Col4";
/* print x; */
/* Afficher les colonnes 1 à 3 */
print (x[1:3, {1 2 3}])[label="Mes données" colname=varNames]; 
print (x[1:3, 1:3])[label="Mes données" colname=varNames];
quit;
```
